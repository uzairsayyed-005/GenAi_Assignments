```python
# importing libraries
import torch
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt
```

```python
# loading pre trained text to image model
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5"
)
```

```python
# giving prompt
prompt = "a futuristic city floating in the sky, digital art, highly detailed"
```

```python
# generating image
image = pipe(prompt).images[0]
```

```python
# displaying image
plt.imshow(image)
plt.axis("off")
```
